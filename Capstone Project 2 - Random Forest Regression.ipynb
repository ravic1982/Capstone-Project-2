```python
df = pd.merge(features_df,stores_df,on='Store',how='inner')
```

```python
df.head()
```

```python
final_features = df
```

```python
dept_count_df = train_data_df.groupby('Store')['Dept'].nunique().reset_index()
```

```python
train_data_df.head()
```

```python
final_df = pd.merge(train_data_df,final_features,on=('Store','Date'),how='inner')
```

```python
final_df
```

```python
a = final_df.groupby(['Store','Date'])['Weekly_Sales'].sum().reset_index()
store_weekly_sales_df = pd.merge(a,final_features,on=('Store','Date'),how='inner')
store_weekly_sales_df
```

```python
store_weekly_sales_df['Date'] = pd.to_datetime(store_weekly_sales_df['Date'])
store_weekly_sales_df['Month'] = store_weekly_sales_df['Date'].dt.month
store_weekly_sales_df['dt_date'] = store_weekly_sales_df['Date'].dt.day
```

```python
store_weekly_sales_df = pd.merge(store_weekly_sales_df,dept_count_df,on=('Store'),how='left')
```

```python
store_weekly_sales_df
```

```python
store_weekly_sales_df['IsHoliday'] = store_weekly_sales_df['IsHoliday'].apply(lambda x: 1 if x == True else x)
store_weekly_sales_df['IsHoliday'] = store_weekly_sales_df['IsHoliday'].apply(lambda x: 0 if x == False else x)
```

```python
a = pd.get_dummies(store_weekly_sales_df)
```

```python
b = a.drop('Type_C',axis=1)
```

```python
store_weekly_sales_df = b.drop('Date',axis=1)
```

```python
store_weekly_sales_df.head()
```

```python
y = store_weekly_sales_df['Weekly_Sales']
X = store_weekly_sales_df.drop('Weekly_Sales',axis=1).values
```

```python
X
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values = 'NaN',strategy = 'mean',axis = 0)
imp.fit(X)
X = imp.transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = RandomForestRegressor(n_estimators=100, max_features=2, min_samples_split=3)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
sc = clf.score(X_test,y_test)
sc
```
