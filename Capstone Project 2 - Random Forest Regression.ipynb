```python
features_df = pd.read_csv(r'C:\Users\rchauhan\Desktop\Springboard Materials\Capstone Project 2\Walmart Data\features.csv')
stores_df = pd.read_csv(r'C:\Users\rchauhan\Desktop\Springboard Materials\Capstone Project 2\Walmart Data\stores.csv')
train_data_df = pd.read_csv(r'C:\Users\rchauhan\Desktop\Springboard Materials\Capstone Project 2\Walmart Data\train.csv')
test_data_df = pd.read_csv(r'C:\Users\rchauhan\Desktop\Springboard Materials\Capstone Project 2\Walmart Data\test.csv')
sample_submission_df = pd.read_csv(r'C:\Users\rchauhan\Desktop\Springboard Materials\Capstone Project 2\Walmart Data\sampleSubmission.csv')
```

```python
df = pd.merge(features_df,stores_df,on='Store',how='inner')
```

```python
df.head()
```

```python
final_features = df
```

```python
dept_count_df = train_data_df.groupby('Store')['Dept'].nunique().reset_index()
```

```python
train_data_df.head()
```

```python
final_df = pd.merge(train_data_df,final_features,on=('Store','Date'),how='inner')
```

```python
final_df
```

```python
a = final_df.groupby(['Store','Date'])['Weekly_Sales'].sum().reset_index()
store_weekly_sales_df = pd.merge(a,final_features,on=('Store','Date'),how='inner')
store_weekly_sales_df
```

```python
store_weekly_sales_df['Date'] = pd.to_datetime(store_weekly_sales_df['Date'])
store_weekly_sales_df['Month'] = store_weekly_sales_df['Date'].dt.month
store_weekly_sales_df['dt_date'] = store_weekly_sales_df['Date'].dt.day
```

```python
store_weekly_sales_df = pd.merge(store_weekly_sales_df,dept_count_df,on=('Store'),how='left')
```

```python
store_weekly_sales_df
```

```python
store_weekly_sales_df['IsHoliday'] = store_weekly_sales_df['IsHoliday'].apply(lambda x: 1 if x == True else x)
store_weekly_sales_df['IsHoliday'] = store_weekly_sales_df['IsHoliday'].apply(lambda x: 0 if x == False else x)
```

```python
a = pd.get_dummies(store_weekly_sales_df)
```

```python
b = a.drop('Type_C',axis=1)
```

```python
store_weekly_sales_df = b.drop('Date',axis=1)
```

```python
store_weekly_sales_df.head()
```

```python
features = ['Store','Fuel_Price','MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5','CPI','Unemployment','IsHoliday','Size','Month','dt_date','Dept','Type_A','Type_B']
```

```python
store_weekly_sales_df['Weekly_Sales_in_millions'] = store_weekly_sales_df['Weekly_Sales']/1000000
```

```python
y = store_weekly_sales_df['Weekly_Sales_in_millions'].values
X = store_weekly_sales_df[features].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import Imputer
from sklearn.pipeline import Pipeline

imp = Imputer(missing_values = 'NaN',strategy = 'mean',axis = 0)
clf = RandomForestRegressor(n_estimators=100, max_features=2, min_samples_split=3)

steps = [('imputation',imp),('RandomForest',clf)]
pipeline = Pipeline(steps)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

x = mean_absolute_error(y_test,y_pred)
x
```

```python
y = np.mean(y_test)
y
```

```python
x/y
```

```python
from sklearn.metrics import mean_squared_error

a= mean_squared_error(y_test,y_pred)
a
```

```python
plt.figure(figsize=(16,9))
sns.regplot(y_test,y_pred)
#plt.scatter(y_test,y_pred,c='blue', marker='s', label='-1')
plt.xlabel('Actual Target Values in Millions')
plt.ylabel('Predicted Target Values in Millions')
plt.show()
```
